```python
import os
from IPython.display import Audio, display
import estereo

# List files in the 'wav' folder
wav_folder = os.path.join(os.getcwd(), 'wav')
print('Files in wav folder:')
print(os.listdir(wav_folder))
```

```python
# Use the Komm sample audio in the wav folder
input_stereo = os.path.join('wav', 'komm.wav')
output_mono = os.path.join('wav', 'komm_mono.wav')

# Convert the stereo file to mono using the half-sum option (default, canal=2)
estereo.estereo2mono(input_stereo, output_mono, canal=2)
print(f'Converted stereo file {input_stereo} to mono file {output_mono}')
```

```python
# Play the original stereo file
print('Original stereo audio:')
display(Audio(filename=input_stereo))

# Play the converted mono file
print('Converted mono audio:')
display(Audio(filename=output_mono))
```
